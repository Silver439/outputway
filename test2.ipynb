```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, ConstantKernel as C

# 定义高斯过程的核函数，这里使用RBF核函数
kernel = C(1.0, (1e-3, 1e3)) * RBF(1.0, (1e-2, 1e2))

# 定义X范围，可以根据需求进行调整
X = np.linspace(0, 5, 100)[:, np.newaxis]

# 创建高斯过程回归模型
gp = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=10)

# 生成样本路径
np.random.seed(0)
y_samples = gp.sample_y(X, n_samples=5)  # 生成5条样本路径

# 绘制样本路径
plt.figure(figsize=(10, 6))
for i in range(5):
    plt.plot(X, y_samples[:, i], lw=2, label=f'Sample {i+1}')

plt.xlabel('X')
plt.ylabel('Sample Path')
plt.title('Gaussian Process Sample Paths')
plt.legend()
plt.show()
```
